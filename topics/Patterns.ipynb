```python
def p20():
    n=int(input("n = "))
    char='A'
    for i in range(n):
        for j in range(n-i-1):
            print(" ",end=" ")
        
        for j in range(n-i-1,n):
            print(chr(ord(char)+j),end=" ")
        print(" ")
p20()
```

```python
def p23():
    n=int(input(" n ="))
    space=(2*n)-2
    for i in range(1,(2*n)):
        stars=i
        if i>n:
            stars=(2*n)-i
            
        #stars
        for j in range(stars):
            print("*",end=" ")

        # spaces
        for j in range(space):
            print(" ",end=" ")

        #stars
        for j in range(stars):
            print("*",end=" ")
        if i>=n:
            space+=2
        else:
            space-=2
        print(" ")
p23()
```

```python
def p23():
    n = int(input("n = "))
    space = (2 * n) - 2   # move outside loop

    for i in range(1, (2 * n)):

        if i <= n:
            stars = i
        else:
            stars = (2 * n) - i

        # Left stars
        for j in range(stars):
            print("*", end=" ")

        # Spaces
        for j in range(space):
            print(" ", end=" ")

        # Right stars
        for j in range(stars):
            print("*", end=" ")

        # Adjust spaces
        if i < n:
            space -= 2
        else:
            space += 2

        print()

p23()
```
