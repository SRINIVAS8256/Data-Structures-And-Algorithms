```python
def divisor(num):
    for i in range(1,num+1,1):
        if num%i == 0:
            print(i)
divisor(36)   #O(logn)  TC
```

```python
# to optimize T C
import numpy as np
def all_div(num):
    divisors=[]
    sqrt=int(pow(num,0.5))
    # for i in range(1,sqrt+1,1):   or
    for i in range(1,sqrt,1):
        if num%i == 0:
            divisors.append(i)
            if i != num//i:
                divisors.append(num//i)
    return sorted(divisors)
```

```python
all_div(36)
```
