# check for prime number

```python
def pn(num):
    count=0
    sqrt=int(pow(num,0.5))
    for i in range(1,sqrt+1,1):
        if num%i == 0:
            # print(i)
            count=count+1
            if i != num//i:
                # print(num//i)
                count=count+1
    if count == 2:
        return "prime"
    else:
        return "not prime"
pn(9)  
```
