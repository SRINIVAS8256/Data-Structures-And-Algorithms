```python
def armstrong(min_num,max_num):
    for i in range(min_num,max_num,1):
        temp=i
        arm=0
        while(i>0):
            rem=i%10
            arm=arm+ rem ** 3
            i=i//10
        if temp == arm:
            print(temp)
```

```python
armstrong(0,10000)
```

```python
#count armstrong
def count_armstrong(min_num,max_num):
    count=0
    for i in range(min_num,max_num,1):
        temp=i
        arm=0
        while(i>0):
            rem=i%10
            arm=arm+ rem ** 3
            i=i//10
        if temp == arm:
            # print(temp)
            count=count+1
    return count
```

```python
count_armstrong(0,10000)
```
