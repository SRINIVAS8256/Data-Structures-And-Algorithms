```python
# reverse the string
ascii=[]
string="srinivas"
def r(string):
    for i in range(len(string)-1,-1,-1):
        ascii.append(string[i])
    return ''.join(ascii)
r(string)
```

```python
# Linear search
def ls(arr,target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return 0
arr=[1,2,34,54,5456,34,31,8]
target =8
ls(arr,target)
```
