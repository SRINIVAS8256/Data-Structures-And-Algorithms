```python
# brute force approach with lower to higher (1,min(n1,n2))
def gcd(n1,n2):
    for i in range(1,min(n1,n2) + 1,1):
        if n1%i == 0 and n2%i == 0:
            ans = i
    return ans 
gcd(1,1)
```

```python
# brute force approach with higher to lower (min(n1,n2),1,-1)
def gcd(n1,n2):
    for i in range(min(n1,n2) + 1,0,-1):
        if n1%i == 0 and n2%i == 0:
            return i
gcd(1,1)
```
