```python
def palindrome(num):
    temp=num
    ans=0
    while(num>0):
        rem=num%10
        ans=ans*10 + rem
        num=num//10
    if temp == ans:
        return "palindrome"
    else:
        return "not palindrome"
palindrome(13333333333333333333333333333333333333333333331)
```

```python
# if number is negative integer

#checking number negative or not hepls to prevent TLE because finding remainder for negative integer sometimes slower



def neg_palindrome(num):
    negative = False
    if num < 0:
        num=num*-1
        negative=True
    temp=num
    ans=0
    while(num>0):
        rem=num%10
        ans=ans*10 + rem
        num=num//10
    if negative == True:
        ans =ans*-1
    if temp == -1*ans or temp == ans:
        return "palindrome"
    else:
        return "not palindrome"
neg_palindrome(-1313131)
```
